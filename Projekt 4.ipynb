```python
from random import randint
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def fu(x):
    return 5/(2*x**2 + 1)

bars = randint(1,30)
barwidth = 5/bars

integral = quad(fu, 0, 5)
a = integral[0]
area = round(a, 6)

fig = plt.figure(figsize = (10,5))
fig.suptitle("Przedziałów jest " + str(bars) + ", pole ≈ " + str(area))
x = np.arange(0,5.1,0.1)
y = fu(x)

x1 = np.linspace(0,5-barwidth,bars)
ax1 = plt.subplot(1,3,1)
ax1.bar(x1+barwidth/2, fu(x1), color = "lightgreen", edgecolor = 'black', width = barwidth)
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0, top=5.2)
ax1.set_xticks([0,1,2,3,4,5])
ax1.scatter(x1, fu(x1), color = 'darkblue', marker = 'o')

areahigh = 0
for i in range(bars):
    areahigh += barwidth * fu(x1[i])

ah = round(areahigh, 6)
ax1.set_title("Suma I: " + str(ah))
ax1.plot(x,y)

ax2 = plt.subplot(1,3,2)
ax2.plot(x,y)
ax2.set_xlim(left=0)
ax2.set_ylim(bottom=0, top=5.2)
ax2.set_xticks([0,1,2,3,4,5])
ax2.fill_between(x,y,facecolor = 'green', alpha = 1)
ax2.text(1.5,3,r"$f(x) = \frac{5}{2x² + 1}$", size = 12)

x3 = np.linspace(barwidth,5,bars)
ax3 = plt.subplot(1,3,3)
ax3.bar(x3-barwidth/2, fu(x3), color = "lightgreen", edgecolor = 'black', width = barwidth)
ax3.set_xlim(left=0)
ax3.set_ylim(bottom=0, top=5.2)
ax3.set_xticks([0,1,2,3,4,5])
ax3.scatter(x3, fu(x3), color = 'darkblue', marker = 'o')

arealow = 0
for i in range(bars):
    arealow += barwidth * fu(x3[i])

al = round(arealow, 6)
ax3.set_title("Suma II: " + str(al))
ax3.plot(x,y)

plt.savefig('05_lew_275116.pdf',dpi=300)
plt.show()
```
