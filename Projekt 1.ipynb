```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot()

N = 100000
x = np.linspace(-2,6,N)
y = -(x**3 - 6*x**2 + x - 1)
ax.plot(x,y, color = 'darkred')
dydx = np.gradient(x, y)
xExtreme = x[1:][dydx[1:]*dydx[:-1]<0]
yExtreme = y[1:][dydx[1:]*dydx[:-1]<0]
ax.scatter([xExtreme], [yExtreme], color = 'black', marker = 'o')
ax.set_xticks(xExtreme)
plt.annotate('-(x - 6x + x - 1)', xy = (3.25, 32), fontsize = 15)
plt.savefig('01_lew_275116.pdf')
```
